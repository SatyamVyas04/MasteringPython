```python
import pandas as pd
import numpy as np

df = pd.read_csv("./bank.csv", sep=";")
df
```

```python
df.head()
```

```python
df.tail()
```

```python
def replace_marital(val):
  if val == "single":
    return 0
  return 1
df["marital"] = df["marital"].apply(replace_marital, 1)
df.head()
```

```python
df
```

```python
df["housing"] = df["housing"].replace({
    "no": 0,
    "yes": 1
})

df
```

```python
df["loan"] = df["loan"].map({
    "no": 0,
    "yes": 1
}.get)

df
```

```python
df["job"].unique()
```

```python
df["job"].replace({
    "unknown": np.nan,
    "unemployed": 0,
    "services": 1,
    "management": 2,
    "blue-collar": 3,
    "self-employed": 4,
    "technician": 5,
    "entrepreneur": 6,
    "admin.": 7,
    "student": 8,
    "housemaid": 9,
    "retired": 10
}, inplace=True)

df
```

```python
df["education"].unique()
```

```python
df["education"].replace({
    "primary": 0,
    "secondary": 1,
    "tertiary": 2,
    "unknown": np.nan
}, inplace=True)

df
```

```python
df["default"].replace({
    "no": 0,
    "yes": 1
}, inplace=True)

df
```

```python
print(df["balance"].min())
print(df["balance"].max())
```

```python
df["balance"] = df["balance"].apply(lambda v: (v-df["balance"].min()))/(df["balance"].max()-df["balance"].min())

df
```

```python
df.contact.unique()

df.contact.replace({
    "unknown": np.nan,
    "telephone": 0,
    "cellular": 1
}, inplace=True)

df
```

```python
df.month.unique()
```

```python
df.month = df.month.map({
    'oct': 10,
    'may': 5,
    'apr': 4,
    'jun': 6,
    'feb': 2,
    'aug': 8,
    'jan': 1,
    'jul': 7,
    'nov': 11,
    'sep': 9,
    'mar': 3,
    'dec': 12})
df.head()
```

```python
df.poutcome.unique()

df.poutcome = df.poutcome.map({
    "unknown": np.nan,
    "failure": 0,
    "other": 1,
    "success": 2
})

df
```

```python
df.pdays=df.pdays.apply(lambda v:(v-df.pdays.min())/(df.pdays.max()-df.pdays.min()))

df
```

```python
df.y.unique()

df.y.replace({'no': 0, 'yes': 1}, inplace=True)

df
```

```python
df.duration = df.duration.apply(lambda v:(v-df.duration.min())/(df.duration.max()-df.duration.min()))

df
```
