## __Data Cleaning, Pre-processing and EDA__

```python
## Importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Loading the dataset

df = pd.read_csv("bank-additional-full.csv", sep = ";")
df.head()
```

```python
## Normalizing the numeric columns

df["age"] = (df["age"] - df["age"].min()) / (df["age"].max() - df["age"].min())
df["duration"] = (df["duration"] - df["duration"].min()) / (df["duration"].max() - df["duration"].min())
df["pdays"] = (df["pdays"] - df["pdays"].min()) / (df["pdays"].max() - df["pdays"].min())
df["previous"] = (df["previous"] - df["previous"].min()) / (df["previous"].max() - df["previous"].min())
df["campaign"] = (df["campaign"] - df["campaign"].min()) / (df["campaign"].max() - df["campaign"].min())
df["emp.var.rate"] = (df["emp.var.rate"] - df["emp.var.rate"].min()) / (df["emp.var.rate"].max() - df["emp.var.rate"].min())
df["cons.price.idx"] = (df["cons.price.idx"] - df["cons.price.idx"].min()) / (df["cons.price.idx"].max() - df["cons.price.idx"].min())
df["cons.conf.idx"] = (df["cons.conf.idx"] - df["cons.conf.idx"].min()) / (df["cons.conf.idx"].max() - df["cons.conf.idx"].min())
df["euribor3m"] = (df["euribor3m"] - df["euribor3m"].min()) / (df["euribor3m"].max() - df["euribor3m"].min())
df["nr.employed"] = (df["nr.employed"] - df["nr.employed"].min()) / (df["nr.employed"].max() - df["nr.employed"].min())
df.head()
```

```python
## Handling categorical columns - job

df["job"].unique()
```

```python
df["job"] = df["job"].map({'housemaid': 1, 'services': 2, 'admin.': 3, 'blue-collar': 4, 'technician': 5, 'retired': 6, 'management': 7, 'unemployed': 8, 'self-employed': 9, 'unknown': 10, 'entrepreneur': 11, 'student': 12})
```

```python
## Handling martial column

df["marital"].unique()
```

```python
df["marital"] = df["marital"].map({'married': 1, 'single': 2, 'divorced': 3, 'unknown': 4})
```

```python
## Handling education

df["education"].unique()
```

```python
df["education"] = df["education"].map({'basic.4y': 1, 'high.school': 2, 'basic.6y': 3, 'basic.9y': 4, 'professional.course': 5, 'unknown': 6, 'university.degree': 7, 'illiterate': 8})
```

```python
## Handling default column

df["default"].unique()
```

```python
df["default"] = df["default"].map({'no': 1, 'unknown': 2, 'yes': 3})
```

```python
## Handling the housing column

df["housing"].unique()
```

```python
df["housing"] = df["housing"].map({'no': 1, 'unknown': 2, 'yes': 3})
```

```python
## Handling the loan column

df["loan"].unique()
```

```python
df["loan"] = df["loan"].map({'no': 1, 'unknown': 2, 'yes': 3})
```

```python
## Contact

df["contact"].unique()
```

```python
df["contact"] = df["contact"].map({'cellular': 1, 'telephone': 2})
```

```python
## Month

df["month"].unique()
```

```python
df["month"] = df["month"].map({'may': 1, 'jun': 2, 'jul': 3, 'aug': 4, 'oct': 5, 'nov': 6, 'dec': 7, 'mar': 8, 'apr': 9, 'sep': 10})
```

```python
## Day of week

df["day_of_week"].unique()
```

```python
df["day_of_week"] = df["day_of_week"].map({'mon': 1, 'tue': 2, 'wed': 3, 'thu': 4, 'fri': 5})
```

```python
## Poutcome

df["poutcome"].unique()
```

```python
df["poutcome"] = df["poutcome"].map({'nonexistent': 1, 'failure': 2, 'success': 3})
```

```python
## y

df["y"].unique()
```

```python
df["y"] = df["y"].map({'no': 0, 'yes': 1})
```

```python
df.head(10)
```

```python
## Save as csv

df.to_csv("bank-additional-full-cleaned.csv", index = False)
```
